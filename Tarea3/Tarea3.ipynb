```python
import numpy 
```

```python
data.insert(loc=1, column = 'Distancia', value = (1/(data['Plx']*10**-3)))
data
```

```python
data.insert(loc=2, column = 'MagAbs', value= (data['Vmag']+5-5*(numpy.log10(data['Distancia']))))
data
```

```python
fig, ax = plt.subplots(figsize=(10,12))

ax.set_xlim(-0.5, 2.5)
ax.set_ylim(15 , -10)
ax.grid()
ax.set_title('H-R Diagram- Hipparcos Catalog')

ax.title.set_fontsize(24)
ax.set_xlabel('Índice de color B-V')
ax.set_ylabel('Magnitud absoluta')
plt.scatter(data['BV'], data['MagAbs'], s=1)
plt.savefig('hr.png')
```
